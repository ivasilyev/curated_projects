```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import ticker as mticker

df = pd.read_csv("temp_bench.csv", header="infer", names=["time_s", "cpu", "celsius"]).set_index(
    "time_s").rename_axis(index="Time")
```

```python
sns.set_theme(style="white")

fig, ax = plt.subplots(figsize=(12, 9))
sns.lineplot(data=df["celsius"].rename("Temperature, C"), color="r", ax=ax)
ax.set_ylabel(ax.yaxis.get_label_text(), color="r")
ax.tick_params(axis="y", colors="r")

ax2 = plt.twinx()
sns.lineplot(data=df["cpu"].rename("CPU overall usage, %"), color="b", ax=ax2)

ax.xaxis.set_major_locator(mticker.FixedLocator(ax.get_xticks().tolist()))
_ = ax.set_xticklabels(pd.to_datetime(ax.get_xticks(), unit="s").strftime('%H:%M:%S'))
ax2.set_ylabel(ax2.yaxis.get_label_text(), color="b")
ax2.tick_params(axis="y", colors="b")

plt.show()
```

```python
plt.clf()
plt.close()
```
